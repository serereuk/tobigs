```python
from sklearn.datasets import load_wine
data = load_wine()
```

```python
X = data["data"]
y = data["target"]
```

```python
X.shape
```

```python
from sklearn.cluster import DBSCAN as dbscan
from sklearn.pipeline import make_pipeline as mpp
from sklearn.preprocessing import StandardScaler 
scale = StandardScaler()
model = model = dbscan(eps = 2.95, min_samples = 36)
pipeline = mpp(scale, model)
predict = pipeline.fit_predict(X)
predict
```

```python
print("label", "db", "y")
print ("0: ", list(predict).count(0), list(y).count(0))
print ("-1: ", list(predict).count(-1), list(y).count(1))
print ("1: ", list(predict).count(1), list(y).count(2))
```
